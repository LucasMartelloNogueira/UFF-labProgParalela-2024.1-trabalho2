```python
import pandas as pd
import matplotlib.pyplot as plt

file = "data.csv"
df = pd.read_csv(file)
print(df)
```

```python
nums = {
    "10^6": 10**6,
    "10^7": 10**7,
    "10^8": 10**8    
}

dataFrames = {key: df[df["n"] == nums[key]] for key in nums}
print(list(dataFrames.keys()))

def obter_tempo_todos_n():

    dataFrames.keys

    for key, _df in dataFrames.items():
        # print(_df)
        threads = _df["threads"].values
        times = _df["time"].values
        # print(f"{key=} / {threads=}")
        # print(f"{key=} / {times=}")
        plt.plot(threads, times)
        # plt.savefig(f"{key}.png")

    plt.xlabel("quantidade threads")
    plt.ylabel("tempos (s)")
    plt.title(f"tempo por threads {list(dataFrames.keys())}")
    plt.legend(dataFrames.keys())
    plt.savefig("tempo por threads.png")

obter_tempo_todos_n()
```

```python
def obter_tempo_cada_n_evolutivo():
    legend_keys = []

    for key, _df in dataFrames.items():
        legend_keys.append(key)
        threads = _df["threads"].values
        times = _df["time"].values
        plt.plot(threads, times)
        plt.xlabel("quantidade threads")
        plt.ylabel("tempos (s)")
        plt.title(f"tempo por threads {legend_keys}")
        plt.legend(legend_keys)
        plt.savefig(f"tempo por threads {legend_keys}.png")

obter_tempo_cada_n_evolutivo()
```

```python
def obter_tempo_cada_n(key, color):
    _df = dataFrames[key]
    threads = _df["threads"].values
    times = _df["time"].values
    plt.plot(threads, times, color=color)
    plt.xlabel("quantidade threads")
    plt.ylabel("tempos (s)")
    plt.legend([key])
    plt.title(f"tempo por threads {key}")
    plt.savefig(f"tempo por threads {key}.png")
    plt.close()

colors = ["blue", "orange", "green"]
for key, color in zip(dataFrames.keys(), colors):
    obter_tempo_cada_n(key, color)
```
